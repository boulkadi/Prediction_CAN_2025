```python
import pandas as pd
import pickle

# Load knockout results
with open('data/knockout_results.pkl', 'rb') as f:
    knockout_results = pickle.load(f)

df_fixture_final_with_teams = knockout_results['final']
print("‚úì Final match data loaded")
```

## Extract Final Results

```python
# Get championship and third-place winners
final_match = df_fixture_final_with_teams[df_fixture_final_with_teams['match_number'] == 'Match 52'].iloc[0]
third_place_match = df_fixture_final_with_teams[df_fixture_final_with_teams['match_number'] == 'Match 51'].iloc[0]

champion = final_match['winner']
home_final = final_match['HomeTeam']
away_final = final_match['AwayTeam']
deuxieme = away_final if champion == home_final else home_final
troisieme = third_place_match['winner']

print(f"Champion: {champion}")
print(f"Runner-up: {deuxieme}")
print(f"Third place: {troisieme}")
```

## Display Podium

```python
print("\n" + "="*50)
print("üèÜ PODIUM FINAL CAN 2025 üèÜ")
print("="*50)
print(f"\nü•á 1ER PLACE  : {champion}")
print(f"ü•à 2√®me PLACE : {deuxieme}")
print(f"ü•â 3√®me PLACE : {troisieme}")
print("\n" + "="*50)
```

## Summary of All Knockout Matches

```python
print("\n" + "="*50)
print("ROUND OF 16")
print("="*50)
round_16 = knockout_results['round_16']
for idx, row in round_16.iterrows():
    print(f"{row['match_number']}: {row['HomeTeam']} vs {row['AwayTeam']} ‚Üí Winner: {row['winner']}")

print("\n" + "="*50)
print("QUARTER-FINALS")
print("="*50)
quarter = knockout_results['quarter']
for idx, row in quarter.iterrows():
    print(f"{row['match_number']}: {row['HomeTeam']} vs {row['AwayTeam']} ‚Üí Winner: {row['winner']}")

print("\n" + "="*50)
print("SEMI-FINALS")
print("="*50)
semi = knockout_results['semi']
for idx, row in semi.iterrows():
    print(f"{row['match_number']}: {row['HomeTeam']} vs {row['AwayTeam']} ‚Üí Winner: {row['winner']}")

print("\n" + "="*50)
print("FINALS")
print("="*50)
final = knockout_results['final']
for idx, row in final.iterrows():
    print(f"{row['match_number']}: {row['HomeTeam']} vs {row['AwayTeam']} ‚Üí Winner: {row['winner']}")
```

## Export Summary

```python
# Create summary dataframe
summary = pd.DataFrame({
    'Place': ['ü•á Champion', 'ü•à Runner-up', 'ü•â Third Place'],
    'Team': [champion, deuxieme, troisieme]
})

# Save summary
summary.to_csv('data/final_podium.csv', index=False)
print("‚úì Saved: data/final_podium.csv")

print("\n" + "="*50)
print("FINAL PODIUM SUMMARY")
print("="*50)
print(summary.to_string(index=False))
print("="*50)
```
