```python
def get_data_for_year(driver, year):
    url = f"https://fr.wikipedia.org/wiki/Coupe_d%27Afrique_des_nations_de_football_{year}"
    year_matches = []

    try:
        driver.get(url)
        print(f"Scraping year {year}...")
    except:
        print(f"Error loading {year}")
        return []

    tables = driver.find_elements(By.TAG_NAME, "table")

    for table in tables:
        try:
            rows = table.find_elements(By.TAG_NAME, "tr")
            if not rows:
                print("pas de rows in year ", year)
                continue

            cells = rows[0].find_elements(By.TAG_NAME, "td")
            if len(cells) < 5:
                print("moin de 5 cells in year ", year)
                continue
                

            score_text = cells[2].text
            score_text = score_text.replace("–", "-")
            if "-" not in score_text:
                print("no - ",score_text ,year)
                continue
                

            date = cells[0].text.strip()
            team_A = cells[1].text.strip()
            team_B = cells[3].text.strip()
            stadium = cells[4].text.strip()

            goals_A, goals_B = score_text.split("-")

            year_matches.append([
                year,
                date,
                team_A,
                team_B,
                goals_A.strip(),
                goals_B.strip(),
                stadium
            ])

        except:
            print("Error processing a table in year ", year)
            continue

    return year_matches
```

```python
from scipy.stats import poisson
```

```python
def predict_points(home, away):
    if home in df_team_strength.index and away in df_team_strength.index:
        # goals_scored * goals_conceded
        lamb_home = df_team_strength.at[home,'GoalsScored'] * df_team_strength.at[away,'GoalsConceded']
        lamb_away = df_team_strength.at[away,'GoalsScored'] * df_team_strength.at[home,'GoalsConceded']
        prob_home, prob_away, prob_draw = 0, 0, 0
        for x in range(0,11): #number of goals home team
            for y in range(0, 11): #number of goals away team
                p = poisson.pmf(x, lamb_home) * poisson.pmf(y, lamb_away)
                if x == y:
                    prob_draw += p
                elif x > y:
                    prob_home += p
                else:
                    prob_away += p
        
        points_home = 3 * prob_home + prob_draw
        points_away = 3 * prob_away + prob_draw
        return (float(points_home), float(points_away))
    else:
        return (0, 0)
```

```python
# Fonction complète pour remplacer tous les placeholders
def replace_all_placeholders(fixture_df, group_results, qualified_thirds, dict_table):
    df_replaced = fixture_df.copy()
    available_thirds = dict(qualified_thirds)  # Copie pour pouvoir modifier
    
    # D'abord, gérer les 3èmes places: trier par nombre d'options (moins d'options en premier)
    third_place_matches = []
    for index, row in df_replaced.iterrows():
        for col in ['HomeTeam', 'AwayTeam']:
            cell_value = str(row[col])
            if '3e du groupe' in cell_value:
                groups_in_option = [g.strip() for g in cell_value.split('3e du groupe ')[1].split('/')]
                third_place_matches.append((index, col, groups_in_option))
    
    # Trier par nombre d'options croissant (les plus restrictifs d'abord)
    third_place_matches.sort(key=lambda x: len(x[2]))
    
    # Traiter les 3èmes places dans cet ordre
    for index, col, groups_in_option in third_place_matches:
        rank_to_remove = None
        for rank, info in available_thirds.items():
            if info['group'] in groups_in_option:
                df_replaced.at[index, col] = info['team']
                rank_to_remove = rank
                break
        
        if rank_to_remove is not None:
            del available_thirds[rank_to_remove]
    
    # Ensuite, traiter les 1er et 2ème du groupe
    for index, row in df_replaced.iterrows():
        for col in ['HomeTeam', 'AwayTeam']:
            cell_value = str(row[col])
            
            if '1er du groupe' in cell_value and '3e' not in cell_value:
                group_letter = cell_value.split('1er du groupe ')[1]
                group_key = f'Group_{group_letter}'
                if group_key in group_results:
                    df_replaced.at[index, col] = group_results[group_key]['1er']
            
            elif '2e du groupe' in cell_value:
                group_letter = cell_value.split('2e du groupe ')[1]
                group_key = f'Group_{group_letter}'
                if group_key in group_results:
                    df_replaced.at[index, col] = group_results[group_key]['2ème']
    
    return df_replaced

# Appliquer la fonction à tous les matchs éliminatoires
df_fixture_8_with_teams = replace_all_placeholders(
    df_fixture_8, 
    group_results, 
    qualified_thirds, 
    dict_table
)
```

```python
# Fonction pour déterminer le vainqueur de chaque match
def get_winner(df_fixture_updated):
    """
    Prédit le vainqueur de chaque match basé sur les probabilités
    """
    for index, row in df_fixture_updated.iterrows():
        home, away = row['HomeTeam'], row['AwayTeam']
        points_home, points_away = predict_points(home, away)
        if points_home > points_away:
            winner = home
        else:
            winner = away
        df_fixture_updated.at[index, 'winner'] = winner
    return df_fixture_updated

# Ajouter une colonne 'winner' au dataframe des huitièmes
df_fixture_8_with_teams['winner'] = ''

# Appliquer la fonction
df_fixture_8_with_teams = get_winner(df_fixture_8_with_teams)

print("Huitièmes de finale avec gagnants prédits:")
print(df_fixture_8_with_teams[['HomeTeam', 'AwayTeam', 'match_number', 'winner']])
```

```python
def update_table(df_fixture_round_1, df_fixture_round_2):
    """
    Remplace les placeholders "Vainqueur Match X" par les vrais gagnants
    du tour précédent et initialise la colonne winner pour le prochain tour
    """
    df_updated = df_fixture_round_2.copy()
    
    # Pour chaque match du tour précédent
    for index, row in df_fixture_round_1.iterrows():
        winner = row['winner']
        match_num = row['match_number'].replace('Match ', '')  # Extraire le numéro
        placeholder = f'Vainqueur Match {match_num}'
        
        # Remplacer le placeholder dans HomeTeam et AwayTeam
        df_updated['HomeTeam'] = df_updated['HomeTeam'].replace(placeholder, winner)
        df_updated['AwayTeam'] = df_updated['AwayTeam'].replace(placeholder, winner)
    
    # Initialiser la colonne winner pour le prochain tour
    df_updated['winner'] = ''
    
    return df_updated
```

```python
def update_final(df_fixture_round_1, df_fixture_final):
    """
    Remplace les placeholders "Vainqueur Match X" et "Perdant Match X" 
    par les vrais gagnants et perdants du tour précédent (semi-finales)
    et initialise la colonne winner pour les finales
    """
    df_updated = df_fixture_final.copy()
    
    # Pour chaque match du tour précédent
    for index, row in df_fixture_round_1.iterrows():
        winner = row['winner']
        home_team = row['HomeTeam']
        away_team = row['AwayTeam']
        
        # Déterminer le perdant
        loser = away_team if winner == home_team else home_team
        
        match_num = row['match_number'].replace('Match ', '')  # Extraire le numéro
        winner_placeholder = f'Vainqueur Match {match_num}'
        loser_placeholder = f'Perdant Match {match_num}'
        
        # Remplacer les placeholders dans HomeTeam et AwayTeam
        df_updated['HomeTeam'] = df_updated['HomeTeam'].replace(winner_placeholder, winner)
        df_updated['HomeTeam'] = df_updated['HomeTeam'].replace(loser_placeholder, loser)
        
        df_updated['AwayTeam'] = df_updated['AwayTeam'].replace(winner_placeholder, winner)
        df_updated['AwayTeam'] = df_updated['AwayTeam'].replace(loser_placeholder, loser)
    
    # Initialiser la colonne winner pour les finales
    df_updated['winner'] = ''
    
    return df_updated
```
